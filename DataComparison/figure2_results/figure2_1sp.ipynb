```python
def plot_table(blind_parity_table_cl, blind_parity_table_cl_err,\
              blind_parity_table_sr, blind_parity_table_sr_err, axis_nm):
    # plot
    fig, ax = plt.subplots(1,2,figsize=(8,3))

    # client
    ax[0].set_title('parity, init state  (Client) at '+ axis_nm)
    X,Y = np.meshgrid(np.linspace(0, 3, 4).astype(int), np.linspace(4, 0, 5).astype(int))

    cmap = mpl.cm.seismic
    c = ax[0].pcolormesh(X, Y, blind_parity_table_cl, vmin=-2.0, vmax=2.0, cmap=cmap)
    # fig.colorbar(  c, ax = ax[0])

    for r in range(4):
        for c in range(3):
            if(blind_parity_table_cl_err[r,c] == -1):
                ax[0].text(c + 0.2, 4-r-0.6,\
                           'not \n measured', \
                   fontsize = 10) 
            else:
                ax[0].text(c + 0.2, 4-r-0.6,\
                           str(blind_parity_table_cl[r,c]) + '+/-\n' + str(blind_parity_table_cl_err[r,c]), \
                   fontsize = 10) 
    ax[0].set_xticks(np.array([0.5, 1.5, 2.5]) )
    ax[0].set_xticklabels(['X', 'Y', 'Z'])

    ax[0].set_yticks(np.array([0.5, 1.5, 2.5, 3.5]) )
    ax[0].set_yticklabels(['$\phi=3\pi/2$', '$\phi=\pi$', '$\phi=\pi/2$', '$\phi=0$'])


    # server
    ax[1].set_title('parity, init state (Server) at '+ axis_nm)
    X,Y = np.meshgrid(np.linspace(0, 3, 4).astype(int), np.linspace(4, 0, 5).astype(int))

    cmap = mpl.cm.seismic
    c = ax[1].pcolormesh(X, Y, blind_parity_table_sr, vmin=-2.0, vmax=2.0, cmap=cmap)
    # fig.colorbar(  c, ax = ax[1])

    for r in range(4):
        for c in range(3):
            if(blind_parity_table_sr_err[r,c] == -1):
                ax[1].text(c + 0.2, 4-r-0.8,\
                           'not \n measured', \
                   fontsize = 10) 
            else:
                ax[1].text(c + 0.2, 4-r-0.8,\
                           str(blind_parity_table_sr[r,c]) + '+/-\n' + str(blind_parity_table_sr_err[r,c]), \
                   fontsize = 10) 
    ax[1].set_xticks(np.array([0.5, 1.5, 2.5]) )
    ax[1].set_xticklabels(['X', 'Y', 'Z'])
    ax[1].set_yticks(np.array([0.5, 1.5, 2.5, 3.5]) )
#     ax[1].set_yticklabels(['$\phi=3\pi/2$', '$\phi=\pi$', '$\phi=\pi/2$', '$\phi=0$'])
    ax[1].set_yticklabels(['', '', '', ''])

    plt.show()
```

# fidelity

```python
phase_list = np.array([0, 0.5, 1, 1.5]) * np.pi
x_cor_phase_list = [np.cos(phase_list), -np.cos(phase_list), -np.sin(phase_list), np.sin(phase_list), \
                    0*np.sin(phase_list), 0*np.sin(phase_list)]
y_cor_phase_list = [-np.sin(phase_list), np.sin(phase_list), -np.cos(phase_list), np.cos(phase_list), \
                    0*np.sin(phase_list), 0*np.sin(phase_list)]
z_cor_phase_list = [np.ones(phase_list.shape), np.ones(phase_list.shape), np.ones(phase_list.shape),\
                   np.ones(phase_list.shape), np.ones(phase_list.shape), np.ones(phase_list.shape)]



def cal_fid(blind_parity_table_allbasis, blind_parity_table_allbasis_err, ax_list_for_plot):
    fid_list, fid_list_err = [], []
    for i, ax in enumerate(ax_list_for_plot):
        if(i < 4):
            par_x = blind_parity_table_allbasis[i].T[0]* x_cor_phase_list[i]
            par_y = blind_parity_table_allbasis[i].T[1]* y_cor_phase_list[i]
            par_avg = ( np.sum(par_x) + np.sum(par_y) ) / 4
            
            par_x_err = blind_parity_table_allbasis_err[i].T[0]* x_cor_phase_list[i]
            par_y_err = blind_parity_table_allbasis_err[i].T[1]* y_cor_phase_list[i]
            par_avg_err = ( np.sum(par_x_err**2) + np.sum(par_y_err**2) )**.5 / 4

        elif(i == 4):
            par_avg = np.mean( blind_parity_table_allbasis[i].T[2] )
            par_avg_err =( np.sum( blind_parity_table_allbasis_err[i].T[2]**2 ) )**.5/4
        elif(i == 5):
            par_avg = -np.mean( blind_parity_table_allbasis[i].T[2] )
            par_avg_err =( np.sum( blind_parity_table_allbasis_err[i].T[2]**2 ) )**.5/4
            
        fid_avg = (1 + par_avg)/2
        fid_avg_err = par_avg_err/2
        fid_list += [fid_avg]
        fid_list_err += [fid_avg_err]
        
    fid_list = np.array(fid_list)
    fid_list_err = np.array(fid_list_err)
    
    return np.mean(fid_list), (np.sum(fid_list_err**2) )**.5/ fid_list.shape[0], fid_list, fid_list_err
    

    
# calculate the blindness
def cal_blind(blind_parity_table_allbasis, blind_parity_table_allbasis_err, ax_list_for_plot):
    bl_list, bl_err_list = [], []
    for i, ax in enumerate(ax_list_for_plot):
        print(i, ax)
        bl, bl_err = blindness(blind_parity_table_allbasis[i], blind_parity_table_allbasis_err[i],\
                               loginfo=False)
        bl_list += [bl]
        bl_err_list += [bl_err]
    bl_list, bl_err_list = np.array(bl_list), np.array(bl_err_list)
    bl_mean = np.mean(bl_list)
    bl_mean_err = (np.sum(bl_err_list**2))**.5/len(bl_err_list)
    
    return bl_mean, bl_mean_err
    
    
    



    
    
###### Blindness ########
# calculate the blindness
def blindness(blind_parity_table, blind_parity_table_err, myphases_label=None, loginfo=True):
    len_angle = blind_parity_table.shape[0]
    # mixed together
    blind_all = np.mean(blind_parity_table, axis=0)
    blind_all_err = (np.sum(blind_parity_table_err**2, axis=0))**.5 / len_angle
    parity_len_all = np.sum(blind_all**2)**.5
    
    if(loginfo):
        print("blind_all: ", blind_all)
        print("blind_all_err: ", blind_all_err)
    parity_len_all_err = 1/parity_len_all * (np.sum( (blind_all * blind_all_err)**2 ) )**.5
    
    # length on the Bloch Sphere
    parity_len_diff_phase = np.sum(blind_parity_table**2, axis=1)**.5
    parity_len_diff_phase_err = 1/parity_len_diff_phase * \
        (np.sum( (blind_parity_table*blind_parity_table_err)**2, axis=1))**.5
    if(loginfo):
        print("parity length: ", parity_len_diff_phase, "+/-", parity_len_diff_phase_err)
    
    
    
    p_p_diff_phas, p_m_diff_phas = 1/2 * (1 +parity_len_diff_phase), 1/2 * (1 -parity_len_diff_phase)
    p_p_diff_phas_err = parity_len_diff_phase/2
    s_diff_phas, s_diff_phas_err = np.zeros(len_angle), np.zeros(len_angle)
    
    for i in range(len_angle):  
        if(p_p_diff_phas[i] < 1):
            s_diff_phas[i] \
                = -p_p_diff_phas[i] * np.log(p_p_diff_phas[i]) - p_m_diff_phas[i] * np.log(p_m_diff_phas[i])
            s_diff_phas_err[i] = np.abs( np.log(p_p_diff_phas[i]/(1-p_p_diff_phas[i])) * p_p_diff_phas_err[i] )
            
        else:
            s_diff_phas[i] = 0
            s_diff_phas_err[i] = 0
            
    if(loginfo):
        print("S (different angle, in bit): ", s_diff_phas/np.log(2), "+/-", s_diff_phas_err/np.log(2))
        
    # S of all mixed state
    p_p_all, p_m_all = 1/2 * (1 + parity_len_all), 1/2 * (1 - parity_len_all)
    p_p_all_err = 1/2 * parity_len_all_err
    
    s_all = -p_p_all * np.log(p_p_all) - p_m_all * np.log(p_m_all) if(p_p_all < 1) else 0
    s_all_err = np.abs( np.log(p_p_all/(1-p_p_all)) * p_p_all_err ) if(p_p_all < 1) else 0
    
    if(loginfo):
        print("S(all together, in bit): ", s_all/np.log(2), "+/-", s_all_err/np.log(2))
        
        
    # S of different phases set
    S_diff_phase_avg, S_diff_phase_avg_err = np.mean(s_diff_phas), (np.sum(s_diff_phas_err**2))**.5/len_angle

    if(loginfo):
        print("S(different phase, in bit): ", S_diff_phase_avg/np.log(2), "+/-", S_diff_phase_avg_err/np.log(2))
        plt.bar(range(len_angle + 1), np.concatenate( (s_diff_phas/np.log(2), np.array([s_all]/np.log(2)) ) ), \
                    yerr=np.concatenate( (s_diff_phas_err/np.log(2), np.array([s_all_err/np.log(2)]) ) ) ,\
                label='S (different phase)')
        plt.bar(range(len_angle + 1), np.concatenate( (0*s_diff_phas/np.log(2), np.array([s_all/np.log(2)]) ) ), \
                    yerr=np.concatenate( (0*s_diff_phas_err/np.log(2), np.array([s_all_err/np.log(2)]) ) ) , \
                label='S ( all )')
        if(myphases_label is not None):
            mylabel = myphases_label
            mylabel.append('all')
            plt.xticks(range(len_angle + 1), mylabel)
        plt.ylabel('S')
        plt.legend()
        plt.tight_layout()
        plt.show()
        
    # Info leakage
    info = s_all - S_diff_phase_avg
    info_err = (s_all_err**2 + S_diff_phase_avg_err**2)**.5
    
    
    return info, info_err







```

```python
avg_fid, avg_fid_err, fid_list, fid_list_err = \
    cal_fid(blind_parity_table_cl_allbasis, blind_parity_table_cl_allbasis_err, ax_list_for_plot)
print('\033[1m', 'Average fidelity over {+x, -x, +y, -y}: ', avg_fid, '+/-', avg_fid_err)

```
